#  파이썬 함수 (function)
- 함수 : 어떤 입력값을 받아서 내부에서 연산처리 후 해당 결과값을 외부로 출력할 수 있도록 기능하는 것
 (프로그래밍 언어의 꽃)
- 함수에 값을 입력 & 출력 : input & output
- 함수가 생성(선언) -> 함수가 존재 -> 값을 입력 (인자값 / 인수 -> 매개변수)을 넣을 수도 있고, 없을 수도 있다
- 함수를 사용(호출)한다
- 함수의 선언부 $ 호출부의 "인자값 / 인수 -> 매개변수"의 갯수는 기본적으로 같아야!
(어느 특정 한쪽이 상대적으로 갯수가 많거나 적으므로 정상적으로 실행 x)

# 파이썬 자료구조
- 문자열 = "string"
- 정수 = 10
- 실수 =3.14
- 논리형 = True
> 단일값 저장 자료형태 1:1

- 리스트 = [1, 2, 3]
- 튜플 = 리스트와 거의 똑같음 -> 최초의 선언된 형태의 자료구조를 변경할 수 없음! (1, 2) -> 데이터의 추가 혹은 삭제불가
> liter & 반복문
> 리스트 & 튜플 안에 담겨있는 값들이 구체적으로 어떤 값을 의미하는지 알 수 없음!
> 
- 딕셔너리 = 사전 = {"apple": "사과"} -> {"key" : "value"} 이 한쌍을 property = 속성
- 집합 = 수학시간에 배웠던 집합 개념
> 교집합, 합집합, 차집합, 공집합 등등
> 반복되어지는 값을 허용하지 않음!!
> 데이터를 크롤링 값을 찾아올 떄, 반복적으로 입력된 값이 들어오는 경우 => 1회 처리
> 인스타그램 접속 > 데이터 크롤링 > 네일아트 > 사장님 >똑같은 컨텐츠 반복
> 
> 다중값 저장 자료형태 1:3 / 1:7
