-데이터 및 정보의 시각화 : Tableau, PowerBI, Looker Studio -> 정규표현식


-패턴중심 : 일관성을 가지고 있는 형태

1) [0-9] : 숫자를 찾아오겠다 -> \d          d = deciaml = 10진수
2) ^[0-9] : ^ = 캐럿 = 무조건 ~으로 시작한다는 의미
3) [^0-9] : ^이 안에 쓰이면 not ,  숫자가 아닌 것을 찾아라 -> \D
4) [\t\n\r\f\v] : t : tab -> 긴여백 /
                  n : new line -> 개행 : 행바꿈
                  r : carriage return -> 처음으로 돌아가라
                  f : form feed -> 다음번쨰 페이지로 돌아가라
                  v : verticl tab
5) [^\t\n\r\f\v] : white space (페이지, 탭여백, 행간, 개행)문자가 아닌것들만 찾아와라
6) [A-Z] : 영문대문자를 찾아와라
7) [a-z] : 영문소문자를 찾아와라
8) [A-Za-z] : 영문 대소문자를 찾아와라
9) [A-Za-z0-9] : 영문 대소문자 혹은 숫자를 찾아와라
10) [가-힣] : 한글을 찾아와라
11) [가-힣A-Za-z0-9] : 한글과 영어 대.소문자 숫자까지 찾아와라
12) . : 문자 1개를 의미함 -> D.A -> D
13) re.sub(패턴, 바꿀데이터, 원본데이터) -> substitute : 후보선수 = 교체선수
14) ? : 물음표를 중심으로 앞에있는 문자가 0번 또는 1번 표시되는 패턴 (gold14k) (gold) -> 없을수도 있거나 1번있거나
15) * : *를 중심으로 앞에있는 문자가 0번 또는 그 이상 (1번의 제약 x) [A-Z]*
16) + : +를 중심으로 앞에있는 문자가 1번 또는 그 이상 (0번을 허용하지 않음)
17) {n} : 앞에 있는 문자가 n번 반복되는 패턴을 찾아라 "AD{2}A" -> ADDA
18) {m, n} : 앞에 있는 문자가 m번 반복되는 패턴부터 n번 반복되는 패턴까지 (범위)를 찾아라 (범위를 지정할때 띄어쓰기 불가!!)
19) [] : 대괄호 안에 포함된 문자는 허용 하겠다는 패턴
20) - : 범위를 지정함 [a-d] [d-t]
21) search() : 왼쪽부터 패턴에 매칭되는 문자를 찾다가, 최초 찾는 순간 정지
    findall : 문자열 모두를 끝까지 찾음 -> 리스트 형식으로 값을 반한

```python
import requests
from bs4 import BeautifulSoup
res = requests.get("https://davelee-fun.github.io/blog/crawling_stock_example.html")

soup = BeautifulSoup(res.text, "html.parser")
data = soup.select("ul.table-stock.sty1 > li.row_sty")

for item in data :
    company = item.select_one("div.st_name > span.name").get_text(" ",strip=True).replace(" ","").replace("\n", "")
    price = item.select_one("div.st_price > span.price").get_text(" ",strip=True).replace(" ","").replace("\n", "")
    rate = item.select_one("div.st_rate > span.icon ").get_text(" ",strip=True).replace(" ","").replace("\n", "")
    
    print(company, price, rate)
```
