```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager
from urllib.request import urlretrieve
import time

URL = "https://www.ssg.com"

service = Service(ChromeDriverManager().install())
options = Options()

#options.add_argument("--headless")
#options.add_argument("--disable-gpu")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--window-size=1920,1080")
options.add_argument("--start-maximized")
options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36")
options.add_argument("--lang=ko-KR")
options.add_argument("--no-sandbox")

driver = webdriver.Chrome(service=service, options=options)
#쓱닷컴 사이트열기
driver.get(URL)
time.sleep(4)


#신세계몰 클릭
to_ssgmall = driver.find_element( By.CSS_SELECTOR,"a[href='https://shinsegaemall.ssg.com']")
to_ssgmall.click()
time.sleep(4)
#초특가 클릭
to_sale = driver.find_element(By.CSS_SELECTOR, "a.menu_lnk")
to_sale.click()
time.sleep(4)

#식품 클릭
to_food = driver.find_element(By.ID, "tabs-:Rcledaal6nj6:--tab-8")
to_food.click()
time.sleep(4)

product = driver.find_elements(By.CSS_SELECTOR,"div.css-1sesc2w")


# 50개 상품 : 상품명 + 판매금액 + 할인률 + URL
for idx, product in enumerate(product) :
    name = product.find_elements(By.CSS_SELECTOR, "p.chakra-text css-19bfb2a")
    price = product.find_elements(By.CSS_SELECTOR, "em.css-1oiygnj") 
    sale_rate = product.find_elements(By.CSS_SELECTOR, "em.css-aywnvu")
    url = product.find_elements(By.CSS_SELECTOR, "a.chakra-link css-1j5cjm")

    print(f"{name}-{price}-{sale_rate}-{url}")
    
    

driver.quit()
```
