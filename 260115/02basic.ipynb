# 파이썬 기초문법 
- 출력: print()  내장함수 =bulit in function / 내장: 파이썬 탄생부터 존재
- 입력: input("당신의 이름은")  사용자로부터 어떤 값을 입력받을 수 있음
                  ㄴ plachholder= 입력값의 정보
  - 해당 함수를 통해서 수집한 값은 무조건 "문자열"이 된다는 사실 기억! 
- 연산:
  - 사칙연산: + - / *  나머지% , 제곱근**  : 문자간의 연결 연산자로서도 사용 가능!! 
  - 비교연산: >  <  >=작거나 같다  <= 크거거나 같다 == 같다 != 같지 않다 (좌항=우항, 우항의 값을 좌항으로 삽입)

- 변수:
  - variable = 변할 수 있는 값을 넣는 공간 (x) =그릇
  - 변수명 작명: 직관, 명시 (한눈에 알아볼수 있는)/  숫자가 먼저 올수 없음 (중간에 오는건 가능)
  - tip: 스네이크 표기법(case_study) / 카멜 표기법(caseStudy)
      
- 자료형: 변수 안에 담을 수 있는 값의 형태 -> 문자열(string)/ 정수(Int)/ 실수(float) 소수점/ 논리형(boolean)
  - 형변화: 자동형변환: input()
  - 명시적형변환: 수동적으로 형태를 변환
  - 자료형 확인: type()
  - int(): 정수형 변환
  - float(): 문자형 열변환
  - bool(): 논리형 변환
  - list(): 리스트 자료구조 변환
  - dict(): 딕셔너리 자료구조 변환 
  - tuple(): 튜플 자료구조 변환
  - set(): 세트(집합) 자료구조 변환
 
- 문자열: 독립적인 함수 혹은 속성들이 많이 존재
  - count(): 특정 문자열에서 어떤 단어가 몇 번 사용되었는지를 집계하고자 할 때 (대소문자 구분)
  - len(): length = 길이
  - find(): 해당 문자열에서 인자값으로 전달된 문자가 몇번째 인덱스에 위치하는지 찾고자 할 때\
  - replace(): 특정 문자열을 대체하고자 할때
  - index: 문자열안에 각각의 문자는 고유한 숫자를 가지고있다 -> 문자열 왼쪽 숫자시작: 0부터 시작
  - slicing: 문자열안에서 특정구간을 잘라내서 가져오고 싶을때
  - strip(): 특정 문자열 앞,뒤쪽에 여백 혹은 여백으로 인지할수 있는 escape sequence(\n, \t) 문자열을 제거하고자 할때
  - fomat(): 특정문자열을 원하는 위치에 출력할때 쓰는 메서드 함수
- 형식지정자:
  - 특정타입의 값을 출력하고자 할때, 어떤 형태로 값을 찾아와서 출력 결정
  - %s - string
  - %c - character
  - %d - int
  - %f - float

```python
# 반복문 사용

text = "banana"
target = "a"

idx = text.find("e") # 존재하지 않은 문자를 찾아와라 -> -1
idx

# 예외 조항 처리: 어떤 명령문을 실행하고자 할때, 만약 ~가 아니라면, 해라~!!

while idx != -1 :
    print(idx)
    idx = text.find(target, idx+1)
```

```python
text = "python"

print(text)
print(text[0])
print(text[1])
# 어떤 문자열이던, 마지막 문자의 인덱스넘버는 전체 문자열 개수 -1 동일

print(len(text))
print(text[len(text)-1])
print(text[-1])
print(text[-2])

# 사이트 > 리뷰 크롤링 > 댓글: 
```

```python
text = "python"

print(text)
print(text[2])
print(text[2:4]) # 찾아오고자하는 문자열 [start:finish+1]
print(text[2:6])
print(text[2:]) # 슬라이싱에서 마지막 번째 인덱스 생략 = 끝까지를 의미
print(text[2:len(text)]) # 해당 구문을 많이 사용
print(text[2:-1])

```

```python
name = "I have a {}, I have an {}."
print(name)

print(name.format("pen","apple"))
print("I have a {}, I have an {}.".format("pen","apple"))
print("I have an  {1}, I have a {0}.".format("pen","apple")) # 객체 - 메서드 함수 앞에 .이 있으면 메서드 함수
# 문자열 형식 객체 .format()
```

```python
interest = 0.087
print(format(interest, ".2f")) # 독립함수 (문자, 숫자 x) -> 소수점 위치 일지
```

```python
# 파이썬 내장함수 가운데, 소수점 자리를 올리거나, 내리는 독립 기능 X
# 외부 라이브러리 모듈 찾아와서 사용 !!
# 수학객체 라이브러리 모듈

import math

interest = 0.0891
result01 = math.floor(interest) #바닥 => 버리다
print(result01)

result02 = math.ceil(interest) #천장 => 올리다
print(result02)
```

```python
print("I have a %f, I have an %c." %(3.14,'A')) # 엄격한 프로그래밍 언어 -> " ",' ' 파이썬은 상관 X
print("I have a %f, I have an %c." %(3.14, 65))
print("I have a %f, I have an %s." %(3.14,"AB"))
```
