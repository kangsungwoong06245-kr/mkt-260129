# git 사용방법
1) 로컬컴퓨터 (개인 PC): git 응용프로그램 설치
2) 터미널 : git -v 명령어
- 검색바 > cmd: 검정색
3) 만약 없다면 > 구글 크롬 > git 다운로드 후 설치 > Standalone Installer > Git for Windows/x64 Setup
(chocolatey: 대리로 응용프로그램 다운로드 후 설치해주는 중간 프로그램 윈도우용)
4) git의 소유자 확인 & 변경
- 터미널> git config --list
5) github 회원가입: github 계정정보 값으로 변경
- git: 나의 로컬컴퓨터에서 버전 관리 툴
- github: 나의 로컬컴퓨터에서 관리하고 있는 버전을 타인과 공유해서 함께 사용하고자 할때, 혹은 내가 사용하고 있는 디바이스가 2대 이상인 경우, A에서도 데이터관리, B에서도 데이터관리
- repositorty: github > project에 따른 개별 폴더
- user.email: jseunghwan7
- user.name: thdghghk67@naver.com
- 터미널 재접속 > 관리자모드: 전역(global)세팅
- git config --global user.name "jseunghwan7"
- git config --global user.email "thdghghk67@naver.com"
6) git commit & push
- cd "현재나의 루트폴더" 이동 > "C:\Users\user\dataclass_weekday"
- git 초기화: git init
- staging area: 데이터 전송: git add .
- 데이터 traking: git status
- 데이터 labeling: git commit -m "260115"
- 원격 (웹) github: git remote add origin https://github.com/jseunghwan7/mkt-260115.git (레포지토리 주소)
- 로컬(git) & 원격(github) 연결: git remote -v (fetch와 push 둘다 떠야함)
- 로컬 -> 원격으로 데이터 푸쉬: git push origin master

7) git clone & pull
- 당연히 집 컴퓨터 git 설치
- 집 도착 > 컴퓨터: 새로운 폴더(Root 폴더) 
- 해당 루트 폴더: git clone "나의 깃허브 주소" 
- 집에서 학원파일을 오픈해서 추가로 내용 업데이트 > 다시 내 레포지토리 푸쉬
- git add . 해당 명령어부터 시작 (집에서 git init 하지말기)
8) git fork
