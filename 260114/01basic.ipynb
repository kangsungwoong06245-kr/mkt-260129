```python
radius = 10
pi = 3.1415

#변수를 선언해서 다음의 값을 출력 
#원의 지름, 원의 둘레, 원의 넓이 

radius2 = radius * 2
radius3 = 2*pi*radius
radius4 = pi*radius **2
radius5 = pi*radius*radius

print ("지름:", radius2)
print ("원의 둘레:", radius3)
print ("원의 넓이:", radius4)

print(f"원의 넓이: {radius4}") #=>f 스트링 문법/ 최신 문법  

print(radius2, radius3, radius4, radius5)
```
