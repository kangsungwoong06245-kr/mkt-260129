```python
import subprocess
import xml.etree.ElementTree as ET

url = "https://apis.data.go.kr/B551011/GoCamping/basedList?numOfRows=50&pageNo=1&MobileOS=ETC&MobileApp=camping&serviceKey=3443fb00c19bdeca0d810b50acd0e4e657e97793555c42795c1a37524d08f07c&_type=xml&_type=xml"

result = subprocess.run(["curl", "-k", url], capture_output=True, text=True, encoding="utf-8")
root = ET.fromstring(result.stdout)

print(root. findall(".//item"))
```
